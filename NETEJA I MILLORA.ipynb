```python
#MOSTRAR ELS VALORS ÚNICS NO NUMÈRICS
def columnas_no_numericas_valores_unicos(df):
    columnas_no_numericas_valores_unicos = []
    for columna in columnas_no_numericas:
        valores_unicos = df[columna].unique()
        columnas_no_numericas_valores_unicos.append(valores_unicos)
        print(columna, '---->', valores_unicos)
    return columnas_no_numericas_valores_unicos

def columnas_valores_unicos(df):
    columnas_valores_unicos = []
    for columna in df.columns:
        valores_unicos = df[columna].unique()
        columnas_valores_unicos.append(valores_unicos)
        print(columna, '---->', valores_unicos)
    return columnas_no_numericas_valores_unicos

c= columnas_no_numericas_valores_unicos(df)
```

```python
#CANVIAR ELS VALORS ÚNICS NO NUMÈRICS PER VALORS ÚNICS NUMÈRICS
mapeos = {}

# Itera sobre las columnas no numéricas y asigna valores numéricos únicos a cada valor único
for columna in df.select_dtypes(exclude=['float', 'int']).columns:
    valores_originales = df[columna].unique()
    mapeo = {valor: indice for indice, valor in enumerate(valores_originales)}
    mapeos[columna] = mapeo
    df[columna] = df[columna].map(mapeo)

# Ahora, las variables no numéricas se han cambiado por valores numéricos únicos
# Puedes acceder a los mapeos utilizando el diccionario `mapeos`
print("Mapeos:")
for columna, mapeo in mapeos.items():
    print(f"{columna}: {mapeo}")
```

```python
#MIRAR CORRELACIONS
import pandas as pd

# Calcular la matriu de correlació
correlation_matrix = df.corr()

# Seleccionar les característiques amd una correlació alta respectoe al target
target_correlation = correlation_matrix['TARGET'].abs().sort_values(ascending=False)
relevant_features = target_correlation.head(20)  # Excluir el target y seleccionar las 10 primeras
selected_features = relevant_features.index.tolist()
print(selected_features)
print(relevant_features)
```

```python

#ELIMINAR CARECTERÍSTIQUES AMB POCA VARIABILITAT
umbral_varianza = 0.1  # umbral de varianza
varianza_por_columna = df.var()
columnas_con_suficiente_varianza = varianza_por_columna[varianza_por_columna > umbral_varianza].index
df_filtrado = df[columnas_con_suficiente_varianza]

#ELIMINAR CARECTERÍSTIQUES ALTAMENT CORRELACIONADES
umbral_correlacion = 0.9  # umbral de correlación
matriz_correlacion = df_filtrado.corr().abs()
columnas_a_eliminar = set()
for i in range(len(matriz_correlacion.columns)):
    for j in range(i):
        if matriz_correlacion.iloc[i, j] >= umbral_correlacion:
            columna = matriz_correlacion.columns[i]
            columnas_a_eliminar.add(columna)
df_filtrado = df_filtrado.drop(columnas_a_eliminar, axis=1)


```

```python
def nan(df_filtrado, columnas_con_nan, total_filas):
    for columna in columnas_con_nan:
        print("Columna:", columna, "NaN:", df_filtrado[columna].isna().sum(), "NaN %:",(round((df_filtrado[columna].isna().sum() / total_filas) * 100,3)))
    print("                                                                                          ")
```

```python
#ELIMINAR ELS NANS
nan(df_filtrado, columnas_con_nan, total_filas)

#AMT_ANNUITY
df_filtrado = df_filtrado.dropna(subset=['AMT_ANNUITY'])
nan(df_filtrado, columnas_con_nan, total_filas)

#OWN_CAR_AGE
df_filtrado['OWN_CAR_AGE'].head(10)
df_filtrado['OWN_CAR_AGE'] = df_filtrado['OWN_CAR_AGE'].replace(0.0, 1.0)
df_filtrado['OWN_CAR_AGE'] = df_filtrado['OWN_CAR_AGE'].fillna(0.0)
nan(df_filtrado, columnas_con_nan, total_filas)

#CNT_FAM_MEMBERS
df_filtrado['CNT_FAM_MEMBERS'].head(10)
df_filtrado = df_filtrado.dropna(subset=['CNT_FAM_MEMBERS'])
nan(df_filtrado, columnas_con_nan, total_filas)

#OBS_30_CNT_SOCIAL_CIRCLE
df_filtradov2 = df_filtrado.dropna(subset=['OBS_30_CNT_SOCIAL_CIRCLE'])
nan(df_filtradov2, columnas_con_nan, total_filas)
#DEF_30_CNT_SOCIAL_CIRCLE

#DEF_60_CNT_SOCIAL_CIRCLE

#DAYS_LAST_PHONE_CHANGE

#AMT_REQ_CREDIT_BUREAU_MON
#df_filtrado['AMT_REQ_CREDIT_BUREAU_MON'].unique()
df_filtradov2 = df_filtradov2.dropna(subset=['AMT_REQ_CREDIT_BUREAU_MON'])
nan(df_filtradov2, columnas_con_nan, total_filas)

#AMT_REQ_CREDIT_BUREAU_QRT

#AMT_REQ_CREDIT_BUREAU_YEAR


```
