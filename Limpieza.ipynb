```python
mapeos = {}

# Itera sobre las columnas no numricas y asigna valores numricos nicos a cada valor nico
for columna in df.select_dtypes(exclude=['float', 'int']).columns:
    valores_originales = df[columna].unique()
    mapeo = {valor: indice for indice, valor in enumerate(valores_originales)}
    mapeos[columna] = mapeo
    df[columna] = df[columna].map(mapeo)

# Ahora, las variables no numricas se han cambiado por valores numricos nicos
# Puedes acceder a los mapeos utilizando el diccionario `mapeos`
print("Mapeos:")
for columna, mapeo in mapeos.items():
    print(f"{columna}: {mapeo}")
```

```python
# 1. Eliminar caractersticas con poca variabilidad
umbral_varianza = 0.1  # umbral de varianza
varianza_por_columna = df.var()
columnas_con_suficiente_varianza = varianza_por_columna[varianza_por_columna > umbral_varianza].index
df_filtrado = df[columnas_con_suficiente_varianza]

# 2. Eliminar caractersticas altamente correlacionadas
umbral_correlacion = 0.9  # umbral de correlacin
matriz_correlacion = df_filtrado.corr().abs()
columnas_a_eliminar = set()
for i in range(len(matriz_correlacion.columns)):
    for j in range(i):
        if matriz_correlacion.iloc[i, j] >= umbral_correlacion:
            columna = matriz_correlacion.columns[i]
            columnas_a_eliminar.add(columna)
df_filtrado = df_filtrado.drop(columnas_a_eliminar, axis=1)


```

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Separar las caractersticas (variables independientes) de la variable objetivo si es necesario
# features = data.drop('variable_objetivo', axis=1) 

# Separar las caractersticas
X = df_filtradov3.iloc[:, 1:]  # Selecciona todas las columnas excepto la primera (asumiendo que la primera es la variable objetivo)

# Estandarizar las caractersticas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear una instancia de PCA y ajustarla a los datos
pca = PCA()
pca.fit(X_scaled)

# Obtener las componentes principales
principal_components = pca.transform(X_scaled)

# Crear un DataFrame con las componentes principales
components_df = pd.DataFrame(data=principal_components, columns=[f'PC{i}' for i in range(1, pca.n_components_ + 1)])

# Mostrar la varianza explicada por cada componente principal
explained_variance_ratio = pca.explained_variance_ratio_
print("Varianza explicada por cada componente principal:")
print(explained_variance_ratio)

# Graficar la varianza explicada acumulada
plt.plot(np.cumsum(explained_variance_ratio))
plt.xlabel('Nmero de componentes principales')
plt.ylabel('Varianza explicada acumulada')
plt.title('Varianza explicada acumulada por componentes principales')
plt.show()
```
